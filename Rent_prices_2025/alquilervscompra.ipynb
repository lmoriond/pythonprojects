```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("AlquilervsVenta.csv")
data.head()
```

```python
# Configuración del gráfico
plt.figure(figsize=(14, 7))

# Graficar precio de alquiler
plt.plot(data['Mes'], data['Precio eur/m2'], label='Precio Alquiler (eur/m2)', color='blue', marker='o')

# Graficar precio de venta
plt.plot(data['Mes'], data['Precio Venta eur/m2'], label='Precio Venta (eur/m2)', color='orange', marker='o')

# Añadir título y etiquetas
plt.title('Evolución de Precios de Alquiler y Venta (eur/m2)')
plt.xlabel('Fecha')
plt.ylabel('Precio (eur/m2)')
plt.xticks(rotation=45)
plt.grid()
plt.legend()

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

```python
# Crear una figura para graficar
plt.figure(figsize=(14, 12))

# Graficar variación anual de precios de alquiler
plt.plot(data.index, data['Variacion anual %'], label='Variación Anual Alquiler (%)', color='blue', marker='o')

# Graficar variación anual de precios de venta
plt.plot(data.index, data['Variacion % venta anual'], label='Variación Anual Venta (%)', color='orange', marker='o')

# Añadir título y etiquetas
plt.title('Comparación de Variaciones Anuales: Alquiler vs Venta')
plt.xlabel('Fecha')
plt.ylabel('Variación (%)')
plt.xticks(rotation=45)
plt.grid()
plt.legend()

# Mostrar gráfico
plt.tight_layout()
plt.show()
```

```python
# Gráfico de dispersión para variaciones anuales
plt.figure(figsize=(25, 25))
plt.scatter(data['Variacion anual %'], data['Variacion % venta anual'], alpha=0.5)
plt.title('Relación entre Variación Anual del Precio de Alquiler y Venta')
plt.xlabel('Variación Anual Alquiler (%)')
plt.ylabel('Variación Anual Venta (%)')
plt.grid()
plt.tight_layout()
plt.show()
```
