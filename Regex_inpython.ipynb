### Task to learn regular expressions in Python

```python
import re
```

### learn meta characters 
for search at the start of the string and $ to search at the end of the string
^ and dollar sign

```python
st = 'dogs are adorable'
```

```python
p = '^dogs'  # p to create a pattern
```

```python
re.search(p, st)
```

```python
# to see if we found it in the end
p2 = 'dogs$'
```

```python
re.search(p2, st)   # none result
```

## Quantifiers
? + *

```python
st2 = 'color'
st2 = 'colour'
```

```python
re.search('colou?r', st2) # the ? is optional, so both will appear
```

```python
st4 = 'ac'
st5 = 'abc'
st6 = 'abbc'
```

```python
re.search('ab+c', st6) # only 5 and 6 will match
```
