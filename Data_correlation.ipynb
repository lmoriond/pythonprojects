```python
# First let's import the packages we will use in this project
# You can do this all now or as you need them
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) #adjust the configuration of the plots we will create

pd.options.mode.chained_assignment = None



# Now we need to read in the data with pandas
df = pd.read_csv(r'C:\Users\lucas\Downloads\Data_Portfolio\Python_correlation\movies.csv')


# Now let's take a look at the data

df.head()
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns

print(df.dtypes)
```

```python
# Changing data format for budget and gross (.0)

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
# Missing data. There are missing values. I decided to drop all rows that have missing data to avoid the error.

df = df.dropna()

# Extracting the year (is different as the formatting is different. Running the following should extract the correct year.)

df['yearcorrect'] = df['released'].str.extract(pat = '([0-9]{4})').astype(int)
```

```python
# Changing data format for columns = budget and gross (no more ".0")

df['budget'] = df['budget'].astype('int64')

df['gross'] = df['gross'].astype('int64')
```

```python
df.head() # now all changes can be made.
```

```python
# Order the data by gross
df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
# if we want to see all the data and not only a summary = none. 
pd.set_option('display.max_rows', 10)
```

```python
# Check for duplicates in duplicates and DROP them
df.drop_duplicates()
```

```python
# We want to test and compare to see if we have high correlation between columns (budget vs gross ; Company vs gross) with scatterplot

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget For Film')
plt.show()
```

```python
# We use Seaborn to Plot the budget vs gross (we can see a positive trend)

sns.regplot(x='budget', y='gross', data=df, scatter_kws={"color":"red"}, line_kws={"color":"blue"})
```

```python
# Whats the actual correlation? Only for numerical fields HIGH CORRELATION MATRIX

df.corr() #pearson by default (method ='kendall' or 'spearman')
```

```python
# High correlation between budget and gross. how can we visualize it. 1=HIGH CORRELATION 0=NO CORRELATION

correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Are there any Outliers?

df.boxplot(column=['gross'])
```

```python
# Lets see score vs gross

sns.regplot(x="score", y="gross", data=df)
```

```python
# Lets visualize rating vs gross

sns.swarmplot(x="rating", y="gross", data=df)
plt.show()
```
